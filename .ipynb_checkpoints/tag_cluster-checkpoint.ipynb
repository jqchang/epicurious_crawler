```python
import pandas as pd
import numpy as np
import math
import random

df = pd.read_csv('wdbc_data.csv',header=None)
data = df.to_numpy()[:,2:] # omit first 2 columns
labels = df.to_numpy()[:,1]
print(np.unique(labels,return_counts=True))
print(data.shape)
print(labels.shape)
df.head()
```

```python
class kMeans:
    def __init__(self,data,labels,k=2,d=2):
        # k = number of means
        # d = distance power
        print("Initializing k-Means clustering")
        print("k={}, distance=L{}".format(k,d))
        self.data = data
        self.ground_truth = labels
        self.N = data.shape[0]
        self.k = k
        self.d = d
        self.clusters = np.zeros(labels.shape)
        self.centroids = np.empty((k,))
    
    def distance(self,a,b,d=2):
        # a = item 1 (np.array)
        # b = item 2 (np.array)
        x = 0
        for i in range(len(a)):
            x += pow(abs(a[i]-b[i]),d)
        return pow(x,1/d)
    
    def train(self,d=2,max_iter=1000):
        print("Training...")
        self.centroids = self.data[np.random.choice(self.N,self.k,replace=False)]
        new_clusters = np.ones(self.N)
        counter = 0
        while np.any(new_clusters != self.clusters) and counter < max_iter:
            # Assign clusters based on shortest distance
            counter += 1
            self.clusters = new_clusters
            new_clusters = np.array(([np.argmin([self.distance(n,m,d) for m in self.centroids])
                                      for n in self.data]))
            for i in range(self.k):
                belongs_to_centroid = np.array([np.arange(self.N)[new_clusters == i]]).ravel()
                self.centroids[i] = np.mean(self.data[belongs_to_centroid],axis=0)
        if counter < max_iter:
            print("Converged after {} iterations".format(counter))
        else:
            print("Stopped after max {} iterations".format(max_iter))
        
    def check_purity(self):
        for i in range(self.k):
            this_cluster = np.arange(self.N)[self.clusters == i]
            cluster_truth = self.ground_truth[this_cluster]
            classes_found,idx,counts = np.unique(cluster_truth,
                                                 return_index=True,
                                                 return_counts=True)
            print("Cluster {}".format(i+1),classes_found, counts, counts/len(this_cluster))
            
    def avg_dist_to_centroid(self):
        sum_dist = np.zeros((self.k,))
        unique_clust,counts = np.unique(self.clusters,return_counts=True)
        for i in range(self.N):
            this_cluster = self.clusters[i]
            parent_centroid = self.centroids[this_cluster]
            sum_dist[this_cluster] += self.distance(self.data[i],parent_centroid,self.d)
        return sum_dist / counts
    
    def max_dist_to_centroid(self):
        max_dist = np.zeros((self.k,))
        unique_clust,counts = np.unique(self.clusters,return_counts=True)
        for i in range(self.N):
            this_cluster = self.clusters[i]
            parent_centroid = self.centroids[this_cluster]
            dist = self.distance(self.data[i],parent_centroid,self.d) 
            if dist > max_dist[this_cluster]:
                max_dist[this_cluster] = dist
        return max_dist
    
    def max_dist_within_cluster(self):
        max_dist = np.zeros((self.k,))
        for c in range(self.k):
            clust = np.arange(self.N)[self.clusters == c]
            for x in clust:
                for j in clust:
                    dist = self.distance(self.data[x],self.data[j],self.d)
                    if dist > max_dist[c]:
                        max_dist[c] = dist
        return(max_dist)
    
    def print_dist_metrics(self):
        print("Avg distance to each centroid:",end='')
        print(self.avg_dist_to_centroid())
        print("Max distance to each centroid:",end='')
        print(self.max_dist_to_centroid())
        print("Max distance between samples: ",end='')
        print(self.max_dist_within_cluster())
```

### 1. Implement the k-means clustering algorithm. Do not use any inbuilt implementation.

#### a,b,c. Apply the algorithm for k=2, using L1 and L2 distance measures.

##### L1 distance

```python
k2d1 = kMeans(data,labels,k=2,d=1)
k2d1.train()
k2d1.check_purity()
k2d1.print_dist_metrics()
```

##### L2 distance

```python
k2d2 = kMeans(data,labels,k=2,d=2)
k2d2.train()
k2d2.check_purity()
k2d2.print_dist_metrics()
```

- $k=2$ creates the following clusters:
    1. High purity of cluster of the "Malignant" class.
    2. Approximately 80-20 split between "Benign" and "Malignant". By intuition, this is not pure enough to classify a sample as "Benign", as the risk of false negative is very high.
- $k=3$:
    1. Pure cluster of class "Malignant"
    2. 93-6 split of "Benign" and "Malignant" - still not conclusive, but better
    3. 28-71 split of "Benign" and "Malignant" - could be used as a screening test, if sample is in this cluster, then run followup tests
- $k=4$:
    1. Pure cluster of class "Malignant"
    2. Inconclusive cluster (nearly 50:50 split)
    3. 97-2 split of "Benign" and "Malignant" - much closer to pure cluster of "Benign" than before
    4. Nearly pure cluster of class "Malignant"
