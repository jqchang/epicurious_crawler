```python
from keras.applications.vgg16 import VGG16
from keras.preprocessing import image
from keras.preprocessing.image import ImageDataGenerator
from keras.applications.vgg16 import preprocess_input
import numpy as np
import pandas as pd

df = pd.read_csv('food_info_cleaned.csv')
X = df.to_numpy()
T = pd.read_csv('tag_vector.csv').to_numpy()
N = X.shape[0]
paths = X[:,0]
Y = X[:,1:] # calories, carbohydrates, protein, fat

with open('tag_list.csv','r') as f:
    tags = f.readlines()
for i,t in enumerate(tags):
    tags[i] = t[0:-1]
tags = np.array(tags)
tag_counts = list(zip(tags,np.sum(T,axis=0)))
for c in tag_counts:
    print(c)
```

```python
path_df = pd.DataFrame(paths)
model = VGG16(weights='imagenet', include_top=False, input_shape=(620,413,3))
datagen = ImageDataGenerator()
img_generator = datagen.flow_from_dataframe(dataframe=path_df, directory='.',
                                            x_col=0, y_col=0, target_size=(620,413))
features = model.predict_generator(img_generator, verbose=1)

np.savetxt("imgnet.csv", features, delimiter=",")
```
